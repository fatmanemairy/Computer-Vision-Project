#classification models

```python
import os
from random import shuffle
import cv2
import numpy as np
from skimage.feature import hog
from skimage import exposure
from skimage.transform import resize
from tqdm import tqdm
import time

def create_label(i):
    label = np.zeros((20,))
    label[i - 1] = 1
    return label
def extract_hog_features(img):
    # Creating hog features
    fd, hog_image = hog(img, orientations=9, pixels_per_cell=(8, 8),
                        cells_per_block=(2, 2), visualize=True)
    return fd


def load_and_preprocess_test_images(test_folder_path,IMG_SIZ):
    test_data = []
    for i in range(1, 21):
      TEST = TEST_DIR + f'/{i}'
      y = create_label(i)
      for img_name in os.listdir(test_folder_path):
          path = os.path.join(test_folder_path, img_name)
          img_data = cv2.imread(path, 0)
          resized_img = resize(img_data, (IMG_SIZ, IMG_SIZ))  # Assuming IMG_SIZE is 100

          # Apply preprocessing
          preprocessed_img = preprocess_image(resized_img)

          # Extract HOG features
          hog_features = extract_hog_features(preprocessed_img)

          test_data.append([hog_features, None])  # Label is not available for test data
          print(img_name)
    return test_data

def preprocess_image(img):
    # Apply histogram equalization
    img = exposure.equalize_adapthist(img)

    # Apply Gaussian blurring
    img = cv2.GaussianBlur(img, (3, 3), 0)

    return img
def create_test_data(TEST_DIR,IMG_SIZE):
    testing_data = []
    for i in range(1, 21):
        TEST = TEST_DIR + f'/{i}/Validation'
        y = create_label(i)
        for img in tqdm(os.listdir(TEST)):
            path = os.path.join(TEST, img)
            img_data = cv2.imread(path, 0)
            resized_img = resize(img_data, (IMG_SIZE, IMG_SIZE))

            # Apply preprocessing
            preprocessed_img = preprocess_image(resized_img)

            # Creating hog features
            fd, hog_image = hog(preprocessed_img, orientations=9, pixels_per_cell=(8, 8),
                                cells_per_block=(2, 2), visualize=True)
            testing_data.append([fd, y])

    shuffle(testing_data)
    return testing_data

def create_train_data(TRAIN_DIR,IMG_SIZE):
    training_data = []

    for i in range(1, 21):
        TRAIN = TRAIN_DIR + f'/{i}/Train'
        y = create_label(i)

        for img in tqdm(os.listdir(TRAIN)):
            path = os.path.join(TRAIN, img)
            img_data = cv2.imread(path, 0)
            resized_img = resize(img_data, (IMG_SIZE, IMG_SIZE))

            # Apply preprocessing
            preprocessed_img = preprocess_image(resized_img)

            # Creating hog features
            fd, hog_image = hog(preprocessed_img, orientations=9, pixels_per_cell=(8, 8),
                                cells_per_block=(2, 2), visualize=True)

            # Appending both features and label
            training_data.append([fd, y])

    shuffle(training_data)
    return training_data

def sp_and_re_test(test):
    x_test = np.array([i[0] for i in test])
    expected_size_test = x_test.shape[1]
    x_test = x_test.reshape(-1, expected_size_test)
    x_test = x_test.astype('float32') / 255.0
    # Reshape y_test
    y_test = np.array([i[1] for i in test]).reshape(-1, 1)

    return x_test,y_test

def sp_and_re_testdata(test):
    # Reshape test data
    x_test = np.array([i[0] for i in test])
    expected_size_test = x_test.shape[1]
    x_test = x_test.reshape(-1, expected_size_test)
    x_test = x_test.astype('float32') / 255.0
    y_test = np.array([i[1] for i in test])
    return x_test,y_test


def sp_and_re_traindata(train):
    # Reshape and normalize training data
    x_train = np.array([i[0] for i in train])
    expected_size = x_train.shape[1]
    x_train = x_train.reshape(-1, expected_size)
    x_train = x_train.astype('float32') / 255.0
    y_train = np.array([i[1] for i in train])
    return x_train,y_train

########################classification
from sklearn.tree import DecisionTreeClassifier
from sklearn import svm
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np  # Don't forget to import NumPy
from sklearn.neighbors import KNeighborsClassifier
from lightgbm import LGBMClassifier
TRAIN_DIR = 'Data/Product Classification/'
TEST_DIR = 'Data/Product Classification'
IMG_SIZE = 100

# Create or load train and test data
train_data = create_train_data(TRAIN_DIR, IMG_SIZE)
test_data = create_test_data(TEST_DIR, IMG_SIZE)

# split and reshape
x_train, y_train = sp_and_re_traindata(train_data)
x_test, y_test = sp_and_re_testdata(test_data)

# Convert one-hot encoded labels to class indices
y_train = np.argmax(y_train, axis=1)

# Define classifiers
classifiers = {
    'decision_tree': DecisionTreeClassifier(random_state=42),
    'svm': svm.SVC(kernel="linear", C=10000.0, random_state=1000),
    'kneighbors': KNeighborsClassifier(n_neighbors=1, weights='distance', algorithm='brute', leaf_size=25, p=1),
    'LGBM': LGBMClassifier(num_leaves=10, learning_rate=0.005, n_estimators=100
                           , max_depth=8, min_child_samples=5, subsample=0.8, colsample_bytree=0.8, random_state=1000,
                           verbose=-1),

}


# Train and evaluate classifiers
for model_name, classifier in classifiers.items():
    # Train the classifier
    starttraintime=time.time()
    classifier.fit(x_train, y_train)
    endtraintime=time.time()
    training_time = endtraintime - starttraintime

    # Make predictions on the test set
    starttesttime=time.time()
    y_pred = classifier.predict(x_test)
    endtesttime=time.time()
    testin_time=endtesttime - starttesttime
    # Convert one-hot encoded predictions to class indices
    y_test_encoded = np.argmax(y_test, axis=1)

    # Evaluate performance
    accuracy = accuracy_score(y_test_encoded, y_pred)

    classification_rep = classification_report(y_test_encoded, y_pred)
    confusion_mat = confusion_matrix(y_test_encoded, y_pred)

    # Print results
    print(f"\nResults for {model_name}:")
    print(f"Accuracy: {accuracy * 100:.2f}%")
    print("Classification Report:\n", classification_rep)
    print("Confusion Matrix:\n", confusion_mat)

    # Visualize confusion matrix
    plt.figure(figsize=(8, 6))
    sns.heatmap(confusion_mat, annot=True, fmt="d", cmap="Blues", xticklabels=range(10), yticklabels=range(10))
    plt.title(f'Confusion Matrix - {model_name}')
    plt.xlabel('Predicted')
    plt.ylabel('True')
    plt.show()
    print(f"Training time: {training_time:.4f} seconds")
    print(f"Testing time: {testin_time:.4f} seconds")

############Model
from tensorflow import keras
from keras.callbacks import EarlyStopping, ModelCheckpoint
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn import svm



# Create or load train and test data
train_data = create_train_data(TRAIN_DIR, IMG_SIZE)
test_data = create_test_data(TEST_DIR, IMG_SIZE)
# split and reshape
x_train, y_train = sp_and_re_traindata(train_data)
x_test, y_test = sp_and_re_testdata(test_data)

model = keras.Sequential([
    keras.layers.Dense(512, activation='relu', input_shape=(4356,)),
    keras.layers.Dropout(0.2),
    keras.layers.BatchNormalization(),
    keras.layers.Dense(256, activation='relu'),
    keras.layers.Dropout(0.2),
    keras.layers.BatchNormalization(),
    keras.layers.Dense(20, activation='softmax')
])

# Compile the model
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Train the model
checkpoint = ModelCheckpoint("Model.h5", monitor='val_accuracy', verbose=1,
                             save_best_only=True, save_weights_only=False, mode='auto', period=1)
early_stopping = EarlyStopping(monitor='val_accuracy', min_delta=0, patience=20, verbose=1, mode='auto')

history = model.fit(x_train, y_train, epochs=1000, batch_size=32, validation_data=(x_test, y_test), callbacks=[checkpoint, early_stopping])

# Evaluate the model
y_pred_nn = model.predict(x_test)
y_pred_nn = np.argmax(y_pred_nn, axis=1)

# Convert one-hot encoded predictions to class indices
y_test_encoded_nn = np.argmax(y_test, axis=1)

# Calculate confusion matrix
confusion_mat_nn = confusion_matrix(y_test_encoded_nn, y_pred_nn)

# Plot confusion matrix
plt.figure(figsize=(8, 6))
sns.heatmap(confusion_mat_nn, annot=True, fmt="d", cmap="Blues", xticklabels=range(20), yticklabels=range(20))
plt.title('Confusion Matrix - Neural Network')
plt.xlabel('Predicted')
plt.ylabel('True')
plt.show()
model.save("ashmawy.h5")
```
